```python
# 필요 라이브러리 설치
import pandas as pd
```

```python
# 데이터 로드
df = pd.read_csv("/content/drive/MyDrive/태블로 프로젝트/쇼핑몰 유료 검색 분석/final_shop_6modata.csv")
df
```

```python
# 결측값 확인
df.isna().sum()
```

```python
# 중복 데이터 확인
df.duplicated().sum()
```

```python
# 데이터 정보 확인
df.info()
```

```python
# 'Ad Group'의 형태를 string으로 변환
df['Ad Group'] = df['Ad Group'].astype('string')

df.dtypes
```

```python
# 'Ad Group' 형태 분리
df[['Type', 'Match Type', 'Devices', 'Ad Type']] = df['Ad Group'].str.split('-', expand=True)
df
```

```python
# 필요 없는 컬럼 삭제
df.drop(['Ad Group', 'Type'], axis=1, inplace=True)
df
```

```python
# 'Desk' -> 'Desktop', 'Mob' -> 'Mobile' 로 값 대체
df['Devices'] = df['Devices'].replace(to_replace=[" Desk ", " Mob "], value=["Desktop", "Mobile"])
df
```

```python
# 'Ad Type' 컬럼의 값 이름 변경
df['Ad Type'] = df['Ad Type'].replace(to_replace=[' [shop coupon code]', ' [shop discount code]', ' [shop promo]', ' Coupon Code', ' Promo Code', ' [shop coupon]',
                                                  ' [shop promo code]', ' Discount Code', ' Free Shipping', ' Competitor', ' Offer', ' Sale', ' Black Friday/Cyber Monday'], value=["Coupon", "Discount",
                                                    "Promo", "Coupon", "Promo", "Coupon", "Promo", "Discount", "Free Shipping", "Competitor", "Offer", "Sale", "Black Friday/Cyber Monday"])
df
```

```python
# csv 파일로 저장
df.to_csv('df.csv', index=False)
```
